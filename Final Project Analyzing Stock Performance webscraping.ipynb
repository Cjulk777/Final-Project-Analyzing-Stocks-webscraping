#1. Extracting Tesla Stock Data Using yfinance

```python
import yfinance as yf

# Símbolo de Tesla en Yahoo Finance
tesla_symbol = 'TSLA'

# Extraer los datos de acciones de Tesla
tesla_data = yf.download(tesla_symbol, start='2022-01-01', end='2022-12-31')

# Mostrar los primeros registros de los datos de Tesla
print(tesla_data.head())
```

2. Tesla Revenue Webscraping

```python
import requests

# Definir la URL de la API de Alpha Vantage
api_url = 'https://www.alphavantage.co/query'

# Parámetros para la solicitud de datos históricos de precios de acciones
params = {
    'function': 'TIME_SERIES_DAILY',
    'symbol': 'TSLA',
    'apikey': 'TU_API_KEY'  # Reemplazar 'TU_API_KEY' con tu propia clave API de Alpha Vantage
}

# Realiza la solicitud GET a la API de Alpha Vantage
response = requests.get(api_url, params=params)

# Verificar si la solicitud fue exitosa (código de estado 200)
if response.status_code == 200:
    # Convierte la respuesta a formato JSON
    data = response.json()
    
    # Imprimir la respuesta para visualizar los datos
    print(data)
else:
    print("Error al obtener los datos:", response.status_code)
```
